```python
import os
import zipfile

# Define the target directory where the zip file is located
target_dir = r'D:\MySirG\Project\Data Analytics and SQL'

# Define the path to the downloaded zip file
zip_file_path = os.path.join(target_dir, 'orders_data.csv.zip')

# Check if the zip file exists
if os.path.exists(zip_file_path):
    print(f"Found zip file: {zip_file_path}")
    
    # Unzip the file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(target_dir)
        print(f"Extracted contents to: {target_dir}")
    
    # Optionally, delete the zip file after extraction
    os.remove(zip_file_path)
    print(f"Deleted zip file: {zip_file_path}")
else:
    print(f"Zip file not found: {zip_file_path}")
```

The con parameter specifies the connection object, index=False indicates that the DataFrame's index should not be written as a separate column in the database table, and if_exists='replace' indicates that if the table already exists, it should be replaced. Alternatively, you can use if_exists='append' to add new rows to the existing table.

__Find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023

```python
df1['year'] = df1['Order Date'].dt.year

```

```python
df1.head(2)
```

```python
a = df1.groupby(['Month', 'year']).sum('Total_Sale')['Total_Sale'].reset_index()
a
```

```python
pivot_table = a.pivot_table(index='Month', columns='year', values='Total_Sale', aggfunc='sum')
pivot_table
```

```python
#The line pivot_table = pivot_table.reset_index() is used to convert a pandas DataFrame's index back into columns.
pivot_table = pivot_table.reset_index()
pivot_table
```

```python
pivot_table['YoY_growth'] = pivot_table[2022] - pivot_table[2023]
pivot_table

# yoy growth column is added
```
